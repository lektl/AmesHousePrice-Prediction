The dataset contains information from the Ames Assessor’s Office used in computing assessed values for individual residential properties sold in Ames, IA from 2006 to 2010.

## Data cleaning Part 1: Check for missing values

```python
# Drop id and pid columns as they are not needed
train_id = train['id']
train = train.drop(columns=['id','pid'])
```

```python
# Identify features with null values
def plot_null(df):
    if df.isnull().sum().sum() != 0:
        na_df = (df.isnull().sum() / len(df)) * 100      
        na_df = na_df.drop(na_df[na_df == 0].index).sort_values(ascending=True)
        null_data = pd.DataFrame({'Null value %' :na_df})
        null_data.plot(kind = "barh")
        plt.title('Percentage of missing data by feature', fontsize=13)
        plt.show()
    else:
        print('No NAs found')

plot_null(train)
```

```python
# Identify features with zero values
def plot_zero(df):
    zero_df = ((df == 0).sum() / len(df)) * 100      
    zero_df = zero_df.drop(zero_df[zero_df == 0].index).sort_values(ascending=True)
    zero_data = pd.DataFrame({'Zero value %' :zero_df})
    zero_data.plot(kind = "barh")
    plt.title('Percentage of zero value data by feature', fontsize=13)
    plt.show()
    
plot_zero(train)
```

```python
# histogram subplots
def subplot_histograms(dataframe, list_of_columns):
    nrows = int(np.ceil(len(list_of_columns)/4)) 
    fig, ax = plt.subplots(nrows=nrows, ncols=4,figsize=(12, nrows*3)) 
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns): 
        ax[i].hist(dataframe[column],bins=10)
        ax[i].set_title(f'{column} distribution',fontsize=14)
        ax[i].tick_params(labelsize=12)
        ax[i].set_xlabel(column, fontsize=14)
    plt.tight_layout()
```

## Model Prep: Create our features matrix (`X`) and target vector (`y`)

```python
X = train.drop(columns='saleprice')
y = train['saleprice']
```

```python
# Create train/test splits.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Scale our data.
# Relabeling scaled data as "Z" is common.
ss = StandardScaler()
Z_train = ss.fit_transform(X_train)
Z_test = ss.transform(X_test)
```

```python
print(f'Z_train shape is: {Z_train.shape}')
print(f'y_train shape is: {y_train.shape}')
print(f'Z_test shape is: {Z_test.shape}')
print(f'y_test shape is: {y_test.shape}')
```

## Instantiate models

```python
lr = LinearRegression()
ridge_cv = RidgeCV(alphas=np.logspace(0, 5, 100), scoring='r2', cv=5)
lasso_cv = LassoCV(alphas=np.logspace(-3, 0, 100), cv=5, max_iter=50000)
enet = ElasticNetCV(alphas=np.linspace(0.5, 1.0, 100), l1_ratio=0.5, cv=5)
```

## Cross validation

```python
lr_score = cross_val_score(lr, Z_train, y_train, cv=3)
lr_score.mean()
```

```python
# LR estimate test RMSE using cross val
lr_score_RMSE = -cross_val_score(lr, Z_train, y_train, cv=3, scoring= 'neg_root_mean_squared_error').mean()
lr_score_RMSE
```

```python
ridge_cv_score = cross_val_score(ridge_cv, Z_train, y_train, cv=3)
ridge_cv_score.mean()
```

```python
# ridge estimate RMSE
ridge_RMSE_score = -cross_val_score(ridge_cv, Z_train, y_train, cv=3, scoring= 'neg_root_mean_squared_error').mean()
ridge_RMSE_score
```

```python
lasso_cv_score = cross_val_score(lasso_cv, Z_train, y_train, cv=3)
lasso_cv_score.mean()
```

```python
# lasso estimate RMSE
lasso_RMSE_score = -cross_val_score(lasso_cv, Z_train, y_train, cv=3, scoring= 'neg_root_mean_squared_error').mean()
lasso_RMSE_score
```

```python
enet_cv_score = cross_val_score(enet, Z_train, y_train, cv=3)
enet_cv_score.mean()
```

```python
# Elasticnet estimate RMSE
enet_RMSE_score = -cross_val_score(enet, Z_train, y_train, cv=3, scoring= 'neg_root_mean_squared_error').mean()
enet_RMSE_score
```
